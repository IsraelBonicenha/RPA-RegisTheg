```python
# Importação Bibliotecas
import pandas as pd
import pyautogui
import time
```

```python
# Configurações
info_excel = "https://docs.google.com/spreadsheets/etc...", "nome-da-aba"
salvarCada = 10
tempoEntreAcoes = 0.15
tempoEsperaSalvar = 30
modoExecucao = "NC"
```

```python
# Leitura Excel
dt_hashtags = pd.read_excel(info_excel[0],  sheet_name=info_excel[1])
dt_hashtags = dt_hashtags.iloc[:, [0,1]].fillna("")
dt_hashtags.drop_duplicates()
```

```python
# Código do Sistema
if(modoExecucao == "C"):


    pyautogui.PAUSE = tempoEntreAcoes                                     
    pyautogui.hotkey('alt', 'tab')                             
    pyautogui.click(1155, 315)                                

    esperar_salvar = False                                     
    index_marcação = 0                                         

    continue_loop = True                                       
    while continue_loop:                                      

        if esperar_salvar:                                   
            time.sleep(tempoEsperaSalvar)                                                                                      
            pyautogui.press('enter', presses=2)               
            pyautogui.press('tab', presses=4, interval=0.1)   
            pyautogui.hotkey('ctrl', 'down')                  


        for i in range(index_marcação, len(dt_hashtags)):      

            hashtag = dt_hashtags.iloc[i, 0]                   
            valores = dt_hashtags.iloc[i, 1]                   

            pyautogui.write(str(hashtag))                      
            pyautogui.press('right')                           
            pyautogui.write(str(valores))                      

            pyautogui.press('right')                           
            pyautogui.press('space')                           
            pyautogui.press('enter')                           
            pyautogui.press('left', presses=2)                 
            index_marcação += 1                                
            
            if (index_marcação % salvarCada  == 0):                  
                break                                          
            
        
        pyautogui.press('tab', presses=4)                      
        pyautogui.press('enter')                               
        esperar_salvar = True                                  

        if index_marcação >= len(dt_hashtags):                 
            continue_loop = False    
                          
elif (modoExecucao == "NC"):


    pyautogui.PAUSE = tempoEntreAcoes                                   
    pyautogui.hotkey('alt', 'tab')                             
    pyautogui.click(1155, 315)                                 

    esperar_salvar = False                                     
    index_marcação = 0                                         

    continue_loop = True                                       
    while continue_loop:                                      

        if esperar_salvar:                                    
            time.sleep(tempoEsperaSalvar)                                                                                         
            pyautogui.press('enter', presses=2)               
            pyautogui.press('tab', presses=4, interval=0.1)   
            pyautogui.hotkey('ctrl', 'down')                  


        for i in range(index_marcação, len(dt_hashtags)):      

            hashtag = dt_hashtags.iloc[i, 0]                   
            valores = dt_hashtags.iloc[i, 1]                   

            pyautogui.write(str(hashtag))                      
            pyautogui.press('right')                           
            pyautogui.write(str(valores))                      

            pyautogui.press('right')                           
            pyautogui.press('space')                           
            pyautogui.press('Down', presses=2)                 
            pyautogui.press('enter')                           
            pyautogui.press('left', presses=2)                 
            index_marcação += 1                                
            
            if (index_marcação % salvarCada  == 0):                  
                break                                          
            
        
        pyautogui.press('tab', presses=4)                      
        pyautogui.press('enter')                               
        esperar_salvar = True                                  

        if index_marcação >= len(dt_hashtags):                 
            continue_loop = False                              


else:
    print(f"Não tem {modoExecucao}.")
```
